# Zadanie 3

```python
import math

def log2(value):
    return math.log2(value)

def main():
    with open('zad3/data.txt', 'r') as data_file:
        lines = data_file.readlines()
        with open('zad3/result.txt', 'w') as result_file:
            for line in lines:
                result = log2(int(line))
                result_file.write(str(result) + "\n")
                
if __name__ == "__main__":
    main()
```

```python
class Product():
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity
        
class Cart():
    def __init__(self):
        self.list_of_products = [
            Product("Apple", 2.99, 10),
            Product("Banana", 2.59, 11),
            Product("Orange", 3.99, 5)
        ]
        
    def print_list_of_products(self):
        for product in self.list_of_products:
            print(f"Name: {product.name}")
            print(f"Price: {product.price}")
            print(f"Quantity: {product.quantity}")
            print()

    def add(self, name):
        for product in self.list_of_products:
            if name == product.name:
                product.quantity = product.quantity + 1
                
    def remove(self, name):
        for product in self.list_of_products:
            if name == product.name:
                product.quantity = product.quantity - 1
    
    def total_price(self):
        sum = 0
        for product in self.list_of_products:
            sum += product.price * product.quantity
        return sum
             
def main():
    cart = Cart()
    cart.print_list_of_products()
    print(f"Total Price: {cart.total_price()}\n")

    print("After adding a apple")
    cart.add("Apple")
    cart.print_list_of_products()
    print(f"Total Price: {cart.total_price()}\n")

    print("After removing an orange")
    cart.remove("Orange")
    cart.print_list_of_products()
    print(f"Total Price: {cart.total_price()}\n")

if __name__ == "__main__":
    main()
```

# Zadanie 5

```python
class Product():
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity
        
    def __str__(self):
        return f"Name: {self.name}\n" +\
               f"Price: {self.price}\n"
        
class Cart():
    def __init__(self):
        self.list_of_products = [
            Product("Apple", 2.99, 10),
            Product("Banana", 2.59, 11),
            Product("Orange", 3.99, 5)
        ]
    
    def __str__(self):
        return_string = ""
        for product in self.list_of_products:
            return_string += product.__str__()
            return_string += f"Quantity: {product.quantity}\n\n"
        return return_string
    
    def __len__(self):
        sum = 0
        for product in self.list_of_products:
            sum += product.quantity
        return sum
            
    def __iter__(self):
        self.n = 0
        return self
    
    def __next__(self):
        if self.n < self.list_of_products.__len__():
            result = self.list_of_products[self.n]
            self.n += 1
            return result
        else:
            raise StopIteration

    def add(self, name):
        for product in self.list_of_products:
            if name == product.name:
                product.quantity = product.quantity + 1
                
    def remove(self, name):
        for product in self.list_of_products:
            if name == product.name:
                product.quantity = product.quantity - 1
    
    def total_price(self):
        sum = 0
        for product in self.list_of_products:
            sum += product.price * product.quantity
        return sum
                
def main():
    cart = Cart()
    print(cart)
    print(f"Sum of quantity: {cart.__len__()}")
    print(f"Total price: {cart.total_price()}\n")

    print("After adding a apple")
    cart.add("Apple")
    print(cart)
    print(f"Sum of quantity: {cart.__len__()}")
    print(f"Total price: {cart.total_price()}\n")

    print("After removing an orange")
    cart.remove("Orange")
    print(cart)
    print(f"Sum of quantity: {cart.__len__()}")
    print(f"Total price: {cart.total_price()}\n")

    print("All products using implemented __iter__ and __next__")
    for product in cart:
        print(product)
        
if __name__ == "__main__":
    main()
```
